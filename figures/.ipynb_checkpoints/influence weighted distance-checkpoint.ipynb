# Einfluss der Gewichtung der Nachbarn

#### Generate sample data

```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path


from sklearn import neighbors

np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)
T = np.linspace(0, 5, 500)[:, np.newaxis]
y = np.sin(X).ravel()

# Add noise to targets
y[::5] += 1 * (0.5 - np.random.rand(8))
```

#### Import Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper', Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = 1/2.54
```

#### Fit regresssion model

```python
n_neighbors = 5

knn = neighbors.KNeighborsRegressor(n_neighbors, weights='uniform')
y_ = knn.fit(X, y).predict(T)

fig,ax = plt.subplots(figsize=(7.5*cm,5.5*cm))
ax.plot(T, y_, label="prediction")
ax.plot(X, y, '.', label="data")
ax.axis("tight")
ax.legend(loc='lower left')
ax.set_title("A", loc = 'left')
ax.set(xlim=(0,5))

plt.tight_layout()
plt.show()
```

```python
n_neighbors = 5

knn = neighbors.KNeighborsRegressor(n_neighbors, weights='distance')
y_ = knn.fit(X, y).predict(T)

fig,ax = plt.subplots(figsize=(7.5*cm,5.5*cm))
ax.plot(T, y_, label="prediction")
ax.plot(X, y, '.', label="data")
ax.axis("tight")
ax.legend(loc = 'lower left')
ax.set_title("B", loc = 'left')
ax.set(xlim=(0,5))


plt.tight_layout()
plt.show()
```

# kNN Grundstruktur Training im PICASO und Haber-Bosch Bereich

#### Importe

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import ConnectionPatch
from sklearn import neighbors
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score as r2
from sklearn.metrics import max_error
from sklearn.model_selection import GridSearchCV
from pathlib import Path
import joblib
```

#### Import Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper', Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = 1/2.54
```

#### Gleichgewichtsdaten importieren

```python
eq_data_file_haber = Path.cwd().parent / 'data' / 'eq_dataset_x.npz' #Import der GGW Daten HaberBosch
res_haber = np.load(eq_data_file_haber)

eq_data_file_picaso = Path.cwd().parent / 'data' / 'eq_dataset_x_extra.npz' #Import der GGW Daten PICASO
res_picaso = np.load(eq_data_file_picaso)

# Bei Speicherung wurden Daten als T, p, x_0 und x gespeichert
# Inputs T, p, x_0[H2,N2,NH3]
# Output x[H2,N2,NH3]

T = np.concatenate((np.array(res_haber['T'])[:500], np.array(res_picaso['T'])[:500]), 0)
p = np.concatenate((np.array(res_haber['p'])[:500], np.array(res_picaso['p'])[:500]), 0)
x_0 = np.concatenate((np.array(res_haber['x_0'][:500]), np.array(res_picaso['x_0'])[:500]), 0)
x = np.concatenate((np.array(res_haber['x'][:500]), np.array(res_picaso['x'])[:500]), 0)

#print(x_0)

X = np.stack((T, p, x_0[:,0], x_0[:,1], x_0[:,2]), axis = 1)
y = np.stack((x[:,0], x[:,2]), axis = 1) # [x_H2, x_NH3]
#print(X[:,0])

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.8, random_state = 42)

#print(y_train)
```

#### Normalisierung

```python
# Preprocessing Normalisierung der Daten
mean_in = np.mean(X_train, 0) # Mittelwert
std_in = np.std(X_train, 0) # Standardabweichung
mean_out = np.mean(y_train, 0)
std_out = np.std(y_train, 0)

X_train = (X_train - mean_in) / std_in
y_train = (y_train - mean_out) / std_out

X_test = (X_test -mean_in) / std_in
y_test = (y_test - mean_out) / std_out
```

#### Wahl k, Elbow Curve

```python
mae_val_test = []
mae_val_train = []
#mse_val = []
#mre_val = []
for K in range(1,40+1):
    model_test = neighbors.KNeighborsRegressor(n_neighbors = K, weights = 'distance')
    model_test.fit(X_train, y_train)  #fit the model
    pred_test = model_test.predict(X_test) #make prediction on test set
    mae_test = (abs(y_test - pred_test)).mean()
    mae_val_test.append(mae_test)
    
    model_train = neighbors.KNeighborsRegressor(n_neighbors = K, weights = 'distance')
    model_train.fit(X_train, y_train)  #fit the model
    pred_train = model_train.predict(X_train) #make prediction on test set
    mae_train = (abs(y_train - pred_train)).mean()
    mae_val_train.append(mae_train)
#     mse = MSE(y_test,pred) #calculate mse
#     mse_val.append(mse) #store mse values
    #mre = abs((y_test - pred) / y_test).mean() # schlechte Metrik zur Bestimmung von k
    #mre_val.append(mre)
    #print('k = ', K, 'MAE:', mae, 'MSE:', mse)


# best_k_mse = mse_val.index(min(mse_val))
# print('Best k f√ºr MSE:', best_k_mse-1)

best_k_mae = mae_val_test.index(min(mae_val_test)) + 1
print('Best k f√ºr MAE:', best_k_mae)

# fig,ax = plt.subplots(2)
# #ax[2].plot(mre_val, label='MRE')
# ax[0].plot(mse_val, label='MSE')
# ax[0].set(xlabel = '$K$ / 1', ylabel = '$MSE$ /') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
# ax[0].tick_params(direction = 'in') #, length = 20, width = 3)
# ax[1].plot(mae_val, label='MAE')
# ax[1].set(xlabel = '$K$ / 1', ylabel = '$MAE$ /') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
# ax[1].tick_params(direction = 'in') #, length = 20, width = 3)
# plt.title("Elbow Curve")

fig,ax = plt.subplots(figsize=(7.5*cm, 5.5*cm))
#ax[2].plot(mre_val, label='MRE')
ax.plot(np.linspace(1,40, 40),mae_val_test,'--')
#ax.plot(mae_val_train, '-')
ax.set(xlabel = '$k$ / 1', ylabel = 'L1 Loss / 1') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
ax.tick_params(direction = 'in') #, length = 20, width = 3)
ax.grid()
ax.set(xlim=(1,40))
#plt.title("Elbow Curve")

#ax.plot(best_k_mae, mae_val[best_k_mae], '.')
ax.plot(best_k_mae, mae_val_test[best_k_mae-1], marker = 2, markeredgewidth = 0.7, markersize=22,color='black')
ax.text(best_k_mae+0.5, mae_val_test[best_k_mae-1]+0.016, ('$k$ = 4'), ha='center', va='center', fontsize=7, color='black', zorder=100)

#plt.legend()
plt.tight_layout()
plt.show()
```

#### kNN Modell

```python
k = best_k_mae # Anzahl der n√§chsten Nachbarn
knn_model = KNeighborsRegressor(n_neighbors = k, weights = 'distance') # Initialisierung
knn_model.fit(X_train, y_train) # Fitten der Trainingsdaten

score_knn_train = knn_model.score(X_train, y_train)
score_knn_test = knn_model.score(X_test, y_test) # Bestimmtheitsma√ü R^2 der Vorhersage
print('R^2 Train:', score_knn_train)
print('R^2 Test:', score_knn_test) 
```

#### Vorhersage Gleichgewichtsdaten, Error rate

```python
# Vorhersage der Gleichgewichtsdaten
pred_train_norm = knn_model.predict(X_train)
pred_test_norm = knn_model.predict(X_test)

# Denormalisierung
pred_train = pred_train_norm * std_out + mean_out
real_train = y_train * std_out + mean_out
pred_test = pred_test_norm * std_out + mean_out
real_test = y_test * std_out + mean_out

# Performance
# performance_train = pd.DataFrame({ 'True Train Value': real_train, 'Train Prediction': pred_train, 'Train Error': real_train - pred_train})
# performance_test = pd.DataFrame({'True Test Value': real_test, 'Test Prediction': pred_test, 'Test Error': real_test - pred_test})

# print(performance_train)
# print(performance_test)

mae_train = (abs(real_train - pred_train)).mean()
mae_test = (abs(real_test - pred_test)).mean()
mre_train = (abs((real_train - pred_train) / real_train)).mean()
mre_test = (abs((real_test - pred_test) / real_test)).mean()
            
print('Train MAE:', mae_train, 'Train MRE:', mre_train, 'Test MAE:', mae_test, 'Test MRE:', mre_test)
```

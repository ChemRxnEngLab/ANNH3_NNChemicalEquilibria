# Auswirkung von alpha auf XGB-Modell

#### Importe

```python
import torch
from torch import nn
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import r2_score as r2
from sklearn.metrics import max_error
import seaborn as sns
```

#### Laden des Train und Testloss Ã¼ber den Trainingsverlauf

```python
# Standardisierungsdaten
loss_data_file_1 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_011_014.npz'
loss_data_1 = np.load(loss_data_file_1)
loss_data_file_2 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_011_017.npz'
loss_data_2 = np.load(loss_data_file_2)

train_loss_1 = np.array(loss_data_1['train_loss'])
test_loss_1 = np.array(loss_data_1['test_loss'])
train_loss_2 = np.array(loss_data_2['train_loss'])
test_loss_2 = np.array(loss_data_2['test_loss'])
```

#### Importing Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper',Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = cm = 1/2.54
```

#### Plot loss vs epoch

```python
fig,ax = plt.subplots(figsize=(9.75*cm, 6.5*cm))

ax.semilogy(train_loss_1, '-', color = 'darkmagenta', label='$\\alpha$ = 0, train')
ax.semilogy(test_loss_1, '--', color = 'darkmagenta', label='$\\alpha$ = 0, train')
ax.semilogy(train_loss_2, '-', color = 'cadetblue', label='$\\alpha$ = 1, train')
ax.semilogy(test_loss_2, '--', color = 'cadetblue', label='$\\alpha$ = 1, test')

#plt.title("Train and Test Loss")
ax.set(xlabel = 'Rounds / 1', ylabel = 'L1 Loss / 1') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
ax.set(xlim=(0,450))
ax.tick_params(direction = 'in') #, length = 20, width = 3)
ax.set_title('D', loc='left')

plt.legend(ncol = 2)
plt.tight_layout()
plt.show()
```

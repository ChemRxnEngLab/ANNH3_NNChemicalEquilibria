# Einfluss der Größe des Datensatzes auf die Leistung des kNNs

#### Importe

```python
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import numpy as np
from pathlib import Path
```

#### Import Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper', Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = 1/2.54
```

#### Laden der Bestimmtheitsmaße

```python
r2_data_file1 = Path.cwd().parent / 'data' / 'models' / 'metric'/ 'r2_008_016.npz' #Import der GGW Daten
res1 = np.load(r2_data_file1)
r2_data_file2 = Path.cwd().parent / 'data' / 'models' / 'metric'/ 'r2_008_005.npz' #Import der GGW Daten
res2 = np.load(r2_data_file2)
r2_data_file3 = Path.cwd().parent / 'data' / 'models' / 'metric'/ 'r2_008_002.npz' #Import der GGW Daten
res3 = np.load(r2_data_file3)
r2_data_file4 = Path.cwd().parent / 'data' / 'models' / 'metric'/ 'r2_008_003.npz' #Import der GGW Daten
res4 = np.load(r2_data_file4)

# Bei Speicherung wurden Daten als T, p, x_0 und xi gespeichert
# Inputs T, p, x_0[H2,N2,NH3]
# Output xi
r2_500 = np.array(res1['r2_500'])
r2_tsd = np.array(res2['r2_tsd'])
r2_10tsd = np.array(res3['r2_10tsd'])
r2_20tsd = np.array(res4['r2_zwtsd'])

#print(r2_500)
```

#### Zu Array zusammenfügen

```python
r2 = np.stack((r2_500, r2_tsd, r2_10tsd, r2_20tsd), axis=0)
#print(r2)
```

#### Plot R^2 gegen Samples im Datensatz

```python
fig,ax = plt.subplots(figsize=(9.75*cm, 6.5*cm))

ax.plot([500, 1000, 10000, 20000], r2,'d--')
#ax.plot(mae_val_train, '-')
ax.set(xlabel = 'Samples / 1', ylabel = '$R^2$ / 1') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
ax.tick_params(direction = 'in') #, length = 20, width = 3)
ax.grid()
#ax.set(xlim=(1,40))
#plt.title("Datasetsize")

#plt.legend()
plt.tight_layout()
plt.show()
```

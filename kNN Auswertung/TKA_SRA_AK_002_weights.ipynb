# Auswirkung des Optimizers auf das neuronales Netz

#### Importe

```python
import torch
from torch import nn
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import r2_score as r2
from sklearn.metrics import max_error
import seaborn as sns
```

#### Default Datentyp Tensor

```python
torch.set_default_dtype(torch.float64)
```

#### Laden des Train und Testloss Ã¼ber den Trainingsverlauf

```python
# Standardisierungsdaten
loss_data_file_1 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_008_001.npz'
loss_data_1 = np.load(loss_data_file_1)
loss_data_file_2 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_008_005.npz'
loss_data_2 = np.load(loss_data_file_2)

loss_1 = np.array(loss_data_1['mae']) # uniform
loss_2 = np.array(loss_data_2['mae']) # distance
print(loss_1)
```

#### Importing Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper',Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = 1/2.54
```

#### Plot loss vs epoch

```python
fig,ax = plt.subplots(figsize = (9.75*cm,6.5*cm))

ax.plot(np.linspace(1,40, 40),loss_1, '--', color = 'darkmagenta', label='uniform, test ')
ax.plot(np.linspace(1,40, 40),loss_2, '--', color = 'cadetblue', label='distance, test ')

#plt.title("Train and Test Loss")
ax.set(xlabel = 'Epochs / 1', ylabel = 'L1 Loss / 1') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
#ax.set(xlim=(0,200))
ax.tick_params(direction = 'in') #, length = 20, width = 3)
ax.set_title('B', loc='left')
ax.set(xlim=(1,40))

plt.legend()
plt.tight_layout()
plt.show()
```

# Auswirkung der Layeranzahl auf neuronales Netz

#### Importe

```python
import torch
from torch import nn
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import r2_score as r2
from sklearn.metrics import max_error
import seaborn as sns
```

#### Default Datentyp Tensor

```python
torch.set_default_dtype(torch.float64)
```

#### Laden des Train und Testloss Ã¼ber den Trainingsverlauf

```python
# Standardisierungsdaten
loss_data_file_1 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_006_003.npz'
loss_data_1 = np.load(loss_data_file_1)
loss_data_file_2 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_006_002.npz'
loss_data_2 = np.load(loss_data_file_2)
loss_data_file_3 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_005_020.npz'
loss_data_3 = np.load(loss_data_file_3)

train_loss_1 = np.array(loss_data_1['train_loss'])
test_loss_1 = np.array(loss_data_1['test_loss'])
train_loss_2 = np.array(loss_data_2['train_loss'])
test_loss_2 = np.array(loss_data_2['test_loss'])
train_loss_3 = np.array(loss_data_3['train_loss'])
test_loss_3 = np.array(loss_data_3['test_loss'])
```

#### Importing Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper',Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = 1/2.54
```

#### Plot loss vs epoch

```python
fig,ax = plt.subplots(figsize=(9.75*cm, 6.5*cm))

ax.semilogy(train_loss_1, '-', color = 'darkmagenta', label='demirer, train')
ax.semilogy(test_loss_1, '--', color = 'darkmagenta', label='demirer, test')
ax.semilogy(train_loss_2, '-', color = 'cadetblue', label='laloy, train')
ax.semilogy(test_loss_2, '--', color = 'cadetblue', label='laloy, test')
ax.semilogy(train_loss_3, '-', color = 'darkorange', label='best, train')
ax.semilogy(test_loss_3, '--', color = 'darkorange', label='best, test')

#plt.title("Train and Test Loss")
ax.set(xlabel = 'Epochs / 1', ylabel = 'L2 Loss / 1') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
ax.set(xlim=(0,800))
ax.tick_params(direction = 'in') #, length = 20, width = 3)
ax.grid()

plt.legend(ncol = 2)
#plt.legend()
plt.tight_layout()
plt.show()
```

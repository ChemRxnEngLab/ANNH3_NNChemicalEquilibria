# Auswirkung der Layeranzahl auf neuronales Netz

#### Importe

```python
import torch
from torch import nn
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import r2_score as r2
from sklearn.metrics import max_error
import seaborn as sns
```

#### Default Datentyp Tensor

```python
torch.set_default_dtype(torch.float64)
```

#### Laden des Train und Testloss Ã¼ber den Trainingsverlauf

```python
# Standardisierungsdaten
loss_data_file_1 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_005_022.npz'
loss_data_1 = np.load(loss_data_file_1)
loss_data_file_2 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_005_023.npz'
loss_data_2 = np.load(loss_data_file_2)
loss_data_file_3 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_005_020.npz'
loss_data_3 = np.load(loss_data_file_3)
loss_data_file_5 = Path.cwd().parent / 'data' / 'models' / 'hyperparams' / 'loss_005_024.npz'
loss_data_5 = np.load(loss_data_file_5)

train_loss_1 = np.array(loss_data_1['train_loss'])
test_loss_1 = np.array(loss_data_1['test_loss'])
train_loss_2 = np.array(loss_data_2['train_loss'])
test_loss_2 = np.array(loss_data_2['test_loss'])
train_loss_3 = np.array(loss_data_3['train_loss'])
test_loss_3 = np.array(loss_data_3['test_loss'])
train_loss_5 = np.array(loss_data_5['train_loss'])
test_loss_5 = np.array(loss_data_5['test_loss'])
```

#### Importing Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper',Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
```

#### Plot loss vs epoch

```python
fig,ax = plt.subplots()

ax.semilogy(train_loss_1, '-', color = 'darkmagenta', label='1 hidden layer, train ')
ax.semilogy(test_loss_1, '--', color = 'darkmagenta', label='1 hidden layer, test')
ax.semilogy(train_loss_2, '-', color = 'cadetblue', label='2 hidden layer, train ')
ax.semilogy(test_loss_2, '--', color = 'cadetblue', label='2 hidden layer, test')
ax.semilogy(train_loss_3, '-', color = 'darkorange', label='3 hidden layer, train ')
ax.semilogy(test_loss_3, '--', color = 'darkorange', label='3 hidden layer, test')
ax.semilogy(train_loss_5, '-', color = 'crimson', label='5 hidden layer, train ')
ax.semilogy(test_loss_5, '--', color = 'crimson', label='5 hidden layer, test')
#plt.title("Train and Test Loss")
ax.set(xlabel = 'Epochs / 1', ylabel = 'L2 Loss / 1') #Beschriftung Achsen; Kursiv durch $$; Index durch _{}
ax.set(xlim=(0,200))
ax.tick_params(direction = 'in') #, length = 20, width = 3)

plt.legend(ncol = 2)
plt.tight_layout()
plt.show()
```

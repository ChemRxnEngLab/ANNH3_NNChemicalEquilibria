# Parity Plot verschiedene Modelle

#### Importe

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
from pathlib import Path
from sklearn.metrics import r2_score as r2
from sklearn.metrics import max_error
import timeit
from torch import onnx
```

#### Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper', Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = 1/2.54
```

#### Import der vorhergesagten Daten 

```python
ann_data_file = Path.cwd().parent / 'data/models/pred_005_020.npz'
ann_data = np.load(ann_data_file)

x_H2_real_ann = np.array(ann_data['x_H2_real_ann'])
x_H2_pred_ann = np.array(ann_data['x_H2_pred_ann'])

knn_data_file = Path.cwd().parent / 'data/models/pred_008_003.npz'
knn_data = np.load(knn_data_file)
x_H2_real_knn = np.array(knn_data['x_H2_real_knn'])
x_H2_pred_knn = np.array(knn_data['x_H2_pred_knn'])

xgb_data_file = Path.cwd().parent / 'data/models/pred_011_027.npz'
xgb_data = np.load(xgb_data_file)
x_H2_real_xgb = np.array(xgb_data['x_H2_real_xgb'])
x_H2_pred_xgb = np.array(xgb_data['x_H2_pred_xgb'])
```

### Parity Plot

#### Auswertung f√ºr Vorhersage von x_H2 auf den Testdaten

```python
bounds = (0,1)

fig,ax = plt.subplots(figsize =(6.5*cm,6.5*cm))
ax.set_xlim(bounds)
ax.set_ylim(bounds)
# Ensure the aspect ratio is square
ax.set_aspect("equal", adjustable="box")

ax.plot(x_H2_real_ann[100:], x_H2_pred_ann[100:], '.', label = 'aNN', zorder= 4)
ax.plot(x_H2_real_knn[:100], x_H2_pred_knn[:100], '.', label = 'kNN')
ax.plot(x_H2_real_xgb[:100], x_H2_pred_xgb[:100], '.', label = 'XGBoost')
ax.plot([0, 1], [0, 1], "-", color = 'crimson',lw=1 ,transform=ax.transAxes, zorder=0)
ax.plot([bounds[0],bounds[1]], [bounds[0] * 1.1, bounds[1]*1.1], "k--", lw=1)#, label = ('\xb1' '10%')) # Error line
ax.plot([bounds[0],bounds[1]], [bounds[0] * 0.9, bounds[1]*0.9], "k--",lw=1) # Error line
ax.text(0.59, 0.88, '+10%', fontsize=7)
ax.text(0.7, 0.55, '-10%', fontsize=7)
ax.set(xlabel = '$x\mathregular{_{H_2}}$ / 1', ylabel = '$x\mathregular{_{H_2,pred}}$ / 1')
ax.tick_params(direction = 'in') #, length = 20, width = 3)
ax.legend()

#inset axes
x1,x2,y1,y2 = 0, 0.2, 0, 0.2 # subregion of origanal image
axin1 = ax.inset_axes([0.56, 0.04, 0.4, 0.4], xlim=(x1,x2), ylim=(y1,y2), xticks=[], yticks=[], xticklabels=[], yticklabels=[])

axin1.plot(x_H2_real_ann, x_H2_pred_ann, '.', markersize = 4)
axin1.plot(x_H2_real_knn[:200], x_H2_pred_knn[:200], '.', markersize = 4)
axin1.plot(x_H2_real_xgb, x_H2_pred_xgb, '.', markersize = 4)
axin1.plot([0, 1], [0, 1], "-", color = 'crimson',lw=1,zorder=1 )
axin1.plot([bounds[0],bounds[1]], [bounds[0] * 1.1, bounds[1] * 1.1], "k--", lw=1)
axin1.plot([bounds[0],bounds[1]], [bounds[0] * 0.9, bounds[1] * 0.9], "k--", lw=1)

rect = (x1, y1, x2 - x1, y2 - y1)
box1= ax.indicate_inset(rect, edgecolor="black", alpha=1,lw=0.7, zorder=100)

cp1 = ConnectionPatch(xyA=(0.2, 0.0), xyB=(0, 0), axesA=ax, axesB=axin1,
                      coordsA="data", coordsB="axes fraction", lw=0.7, ls=":", zorder=100)
cp2 = ConnectionPatch(xyA=(0.2, 0.2), xyB=(0, 1), axesA=ax, axesB=axin1,
                      coordsA="data", coordsB="axes fraction", lw=0.7, ls=":", zorder=100)

ax.add_patch(cp1)
ax.add_patch(cp2)

#plt.legend()
#fig.suptitle("Parity Plot")
plt.tight_layout()
plt.show()
```

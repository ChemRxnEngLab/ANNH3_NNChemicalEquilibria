# Auftragung Vorhersage beste Modelle gegen T

#### Importe

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
from pathlib import Path
from sklearn.metrics import r2_score as r2
from sklearn.metrics import max_error
import timeit
from torch import onnx
```

#### Plot Stylesheet

```python
plt.style.use(['seaborn-v0_8-paper', Path.cwd().parent / 'plotstyle' / 'paper.mplstyle'])
cm = 1/2.54
```

#### Import der vorhergesagten Daten 

```python
ann_data_file = Path.cwd().parent / 'data/models/verlauf/v_ann_005_020.npz'
ann_data = np.load(ann_data_file)

T = np.array(ann_data['T'])
pred_ann = np.array(ann_data['x_NH3_pred'])
real = np.array(ann_data['x_NH3_real'])

knn_data_file = Path.cwd().parent / 'data/models/verlauf/v_knn_008_003.npz'
knn_data = np.load(knn_data_file)
pred_knn = np.array(knn_data['x_NH3_pred'])

xgb_data_file = Path.cwd().parent / 'data/models/verlauf/v_xgb_011_027.npz'
xgb_data = np.load(xgb_data_file)
pred_xgb = np.array(xgb_data['x_NH3_pred'])
```

### Verläufe mit konstant gehaltenen Parametern

#### Auswertung Verlauf x_NH3 über T, p =const, x_0 = const

```python
fig,ax = plt.subplots(figsize=(9.75*cm, 6.5*cm))

ax.plot(T, real, '-', label = 'real')
ax.plot(T[::3], pred_ann[::3], 'd', markeredgewidth = 0.7, markersize = 4.5,label = 'aNN')
ax.plot(T[1:][::3], pred_knn[1:][::3], '.', markeredgewidth = 0.7, markersize = 4.5,label = 'kNN')
ax.plot(T[2:][::3], pred_xgb[2:][::3], '^', markeredgewidth = 0.7, markersize = 3.5,label = 'xgb')

ax.set(xlabel = '$T$ / K', ylabel = '$x\mathregular{_{NH_3}}$ / 1')
ax.tick_params(direction = 'in') #, length = 20, width = 3)
#ax.set_title('p = const, x_0 = const')
ax.set(xlim=(650,850))
ax.legend()
ax.grid()
#ax.legend(['$\\mathregular{R^2}$ = ', r2(xi_real,xi_pred)], markerscale=0)

#plt.legend()
#fig.suptitle("Parity Plot")
plt.tight_layout()
plt.show()
```
